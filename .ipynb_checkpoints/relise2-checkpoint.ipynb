```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
real = pd.read_csv('https://raw.githubusercontent.com/SoloYolo7/RealEstate/main/data.csv', sep='\t')
```

```python
real
```

```python
real['Parking'] = real['Parking'].apply(lambda x: x.capitalize())
```

```python
real.insert(loc=real.columns.to_list().index('Elevator')+1, column='Count_freight_elevator', \
                  value=real['Elevator'].apply(lambda x: int(x[-2]) if 'Груз' in x else 0))
```

```python
real['Elevator'] = real['Elevator'].apply(lambda x: int(x[x.index('(')+1]) if 'Пасс' in x else 0)
real = real.rename(columns={'Elevator': 'Count_pass_elevator'})
```
