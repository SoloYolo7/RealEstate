```python
# Посмотрим, какие существуют значения и их количество в каждой feature

for feat in real.columns[2:-1]:
    print(real[feat].value_counts(dropna=False))
    print()
    print()
```

```python
# Словарь, в котором будут содержаться наши feature и количетсво NaN 
d_nan = {}

for feat in real.columns[2:-1]:
    if real[feat].isna().sum() != 0:
        d_nan[feat] = real[feat].isna().sum()
d_nan
```

```python
# Построим гистрограмму для данного словаря, чтобы визуально увидеть feature и соответствующее им кол-во значений NaN
from matplotlib.ticker import MultipleLocator

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot()

ax.bar(d_nan.keys(), d_nan.values());
ax.grid()
ax.yaxis.set_major_locator(MultipleLocator(base=1500))
plt.xticks(rotation=75)
plt.tight_layout();
```
